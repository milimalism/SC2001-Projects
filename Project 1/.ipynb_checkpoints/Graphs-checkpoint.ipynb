```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter, MultipleLocator
from matplotlib.lines import Line2D
```

```python
# Assuming you have the data in an Excel file named 'data.xlsx' in the same directory
# Load the data from the Excel file into a pandas DataFrame
df_part1 = pd.read_excel('partc1.xlsx')
df_part2 = pd.read_excel('partc2.xlsx')

#limiting dataframe for clearer graph
start_index = 0
end_index = 12
limited_df_part1 = df_part1.iloc[start_index:end_index + 1]
limited_df_part2 = df_part2.iloc[0:14]
limited_df_part2_2 = df_part2.iloc[18:27]
limited_df_part2_3 = df_part2.iloc[28:45]
limited_df_part2_100k = df_part2.iloc[45:60]
limited_df_part2_300k = df_part2.iloc[54:64]
limited_df_part2_600k = df_part2.iloc[64:74]

# Extract the columns
array_size = df_part1['Array Size']
key_comparisons_avg_part1 = df_part1['Key Comparisons']

s = limited_df_part2['S']
key_comparisons_avg_part2 = limited_df_part2['Key Comparisons']
key_comparisons_best_part2 = limited_df_part2['Best Case']
display(limited_df_part2_100k)
```

```python
# Create a scatterplot
#plt.scatter(array_size, key_comparisons_avg)

def format_ticks(value, pos):
    if abs(value) >= 1e9:
        return f"{value:.0e}"  # Display in scientific notation
    else:
        return f"{value:.0f}"  # Display as an integer"

def plot_scatterPlot_part1(df) :
    # Set the figure size (width and height) in inches
    fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the values as needed

    # Extract the columns
    array_size = df['Array Size']
    key_comparisons_avg = df['Key Comparisons']
    key_comparisons_best = df['Best Case']

    # Create a scatterplot
    ax.scatter(array_size, key_comparisons_avg)


    # Create scatterplots for "Key comparisons avg" and "Key comparisons best"
    ax.scatter(array_size, key_comparisons_avg, label='Key comparisons', color='blue')
    ax.scatter(array_size, key_comparisons_best, label='Key comparisons', color='red')

    # Perform polynomial regression for "Key comparisons avg" (quadratic in this example)
    degree = 2  # Choose the degree of the polynomial (e.g., 2 for quadratic)
    coefficients_avg = np.polyfit(array_size, key_comparisons_avg, degree)
    poly_fit_avg = np.poly1d(coefficients_avg)
    x_fit_avg = np.linspace(min(array_size), max(array_size), 100)  # Generate x values for the trendline
    y_fit_avg = poly_fit_avg(x_fit_avg)

    # Perform polynomial regression for "Key comparisons best" (quadratic in this example)
    coefficients_best = np.polyfit(array_size, key_comparisons_best, degree)
    poly_fit_best = np.poly1d(coefficients_best)
    x_fit_best = np.linspace(min(array_size), max(array_size), 100)  # Generate x values for the trendline
    y_fit_best = poly_fit_best(x_fit_best)

    # Plot the polynomial trendlines
    ax.plot(x_fit_avg, y_fit_avg, label='Average-case Key Comparisons Trendline', linestyle='--', color='blue')
    ax.plot(x_fit_best, y_fit_best, label='Best-case Key Comparisons Trendline', linestyle='--', color='red')

    # Apply the custom formatting function to the x and y axes
    formatter = FuncFormatter(format_ticks)
    ax.xaxis.set_major_formatter(formatter)
    ax.yaxis.set_major_formatter(formatter)


    # Add a grid to the background of the graph
    ax.grid(True)

    ''''# Set a custom interval of 500,000 for the x-axis tick marks
    x_interval = 500000
    ax.xaxis.set_major_locator(MultipleLocator(base=x_interval))'''


    # Create custom legend entries with lines but no dots
    legend_elements = [
        Line2D([0], [0], color='blue', lw=2, label='Average-case Key Comparisons'),
        Line2D([0], [0], color='red', lw=2, label='Best-case Key Comparisons')
    ]

    # Add labels, a title, and the custom legend
    ax.set_xlabel('Array Size')
    ax.set_ylabel('Key Comparisons')
    ax.set_title('Scatterplot of Array Size vs. Key Comparisons')
    ax.legend(handles=legend_elements, loc='upper left')


    # Show the plot
    plt.show()
```

```python
plot_scatterPlot_part1(df_part1)
```

```python
plot_scatterPlot_part1(limited_df_part1)
```

```python
def plot_scatterPlot_part2(df):
    
    # Extract the "S" and "Key Comparisons" columns
    S_values = df['S']
    key_comparisons = df['Key Comparisons']
    best_case = df['Best Case']
    
    def custom_format(x, pos):
        return f'{int(x):,}'

    # Create a scatterplot with custom formatting and colors
    plt.figure(figsize=(15, 6))  # Adjust the figure size as needed

    # Plot "Key Comparisons" data points in blue
    plt.scatter(S_values, key_comparisons, alpha=0.5, color='blue', label='Key Comparisons')

    # Plot "Best Case" data points in red
    plt.scatter(S_values, best_case, alpha=0.5, color='red', label='Best Case')

    # Perform polynomial regression (quadratic) for "Key Comparisons"
    degree = 2  # Choose the degree of the polynomial (e.g., 2 for quadratic)
    coefficients_key = np.polyfit(S_values, key_comparisons, degree)
    poly_fit_key = np.poly1d(coefficients_key)
    x_fit_key = np.linspace(min(S_values), max(S_values), 100)  # Generate x values for the trendline
    y_fit_key = poly_fit_key(x_fit_key)

    # Perform polynomial regression (quadratic) for "Best Case"
    coefficients_best = np.polyfit(S_values, best_case, degree)
    poly_fit_best = np.poly1d(coefficients_best)
    x_fit_best = np.linspace(min(S_values), max(S_values), 100)  # Generate x values for the trendline
    y_fit_best = poly_fit_best(x_fit_best)

    # Plot the polynomial trendlines
    plt.plot(x_fit_key, y_fit_key, linestyle='--', color='blue', label='Key Comparisons Trendline')
    plt.plot(x_fit_best, y_fit_best, linestyle='--', color='red', label='Best Case Trendline')

    # Add labels and a title
    plt.xlabel('S')
    plt.ylabel('Values')
    plt.title('Scatterplot of S vs. Key Comparisons and Best Case')

    # Apply the custom formatting function to the x and y axes
    formatter = FuncFormatter(custom_format)
    plt.gca().xaxis.set_major_formatter(formatter)
    plt.gca().yaxis.set_major_formatter(formatter)
    
    # Create custom legend entries with lines but no dots
    legend_elements = [
        Line2D([0], [0], color='blue', lw=2, label='Average-case Key Comparisons'),
        Line2D([0], [0], color='red', lw=2, label='Best-case Key Comparisons')
    ]

    # Show the legend to differentiate data points and trendlines
    plt.legend(handles=legend_elements, loc='upper left')

    # Show the plot
    plt.grid(True)
    plt.show()
```
