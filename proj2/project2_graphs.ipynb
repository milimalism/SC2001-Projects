```python
import pandas as pd
import matplotlib.pyplot as plt

array = pd.read_csv('outputA.csv')
heap = pd.read_csv('outputB.csv')
```

```python
# Extract data from outputA.csv
edges = []
vertices = []
count = 1

best_ag = []
worst_ag = []
aver_ag = []

best_aa = []
worst_aa = []
aver_aa = []

best_ac = []
worst_ac = []
aver_ac = []

for _, row in array.iterrows():

    if count%3 == 1:
        best_ag.append(float(row['graph creating time']))
        best_aa.append(float(row['algorithm time']))
        best_ac.append(int(row['Number of algocomparisons']))

    elif count%3 == 2:
        worst_ag.append(float(row['graph creating time']))
        worst_aa.append(float(row['algorithm time']))
        worst_ac.append(int(row['Number of algocomparisons']))

    elif count%3 == 0:
        aver_ag.append(float(row['graph creating time']))
        aver_aa.append(float(row['algorithm time']))
        aver_ac.append(int(row['Number of algocomparisons']))
        edges.append(int(row['Number of edges']))
        vertices.append(int(row['Number of vertices']))
    
    count += 1
```

```python
# Extract data from outputB.csv
count = 1

best_hg = []
worst_hg = []
aver_hg = []

best_ha = []
worst_ha = []
aver_ha = []

best_hc = []
worst_hc = []
aver_hc = []

for _, row in heap.iterrows():

    if count%3 == 1:
        best_hg.append(float(row['graph creating time']))
        best_ha.append(float(row['algorithm time']))
        best_hc.append(int(row['Number of algocomparisons']))

    elif count%3 == 2:
        worst_hg.append(float(row['graph creating time']))
        worst_ha.append(float(row['algorithm time']))
        worst_hc.append(int(row['Number of algocomparisons']))

    elif count%3 == 0:
        aver_hg.append(float(row['graph creating time']))
        aver_ha.append(float(row['algorithm time']))
        aver_hc.append(int(row['Number of algocomparisons']))
    
    count += 1
```

```python
# Array - Graph creation
print(best_ag)
print(worst_ag)
print(aver_ag)

# Creating the graph
plt.figure(figsize=(10, 6))
plt.plot(edges, best_ag, label = "Best")
plt.plot(edges, worst_ag, label = "Worst")
plt.plot(edges, aver_ag, label = "Average")

plt.legend()
plt.xlabel('Number of Edges')
plt.ylabel('Graph Creating Time')
plt.title('Array: Graph Creation Time vs. Number of Edges')
plt.show()
```

```python
# Array - Algorithm time
print(best_aa)
print(worst_aa)
print(aver_aa)

# Creating the graph
plt.figure(figsize=(10, 6))
plt.plot(edges, best_aa, label = "Best")
plt.plot(edges, worst_aa, label = "Worst")
plt.plot(edges, aver_aa, label = "Average")

plt.legend()
plt.xlabel('Number of Edges')
plt.ylabel('Algorithm Time')
plt.title('Array: Algorithm Time vs. Number of Edges')
plt.show()
```
